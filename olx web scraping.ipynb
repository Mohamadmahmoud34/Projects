```python
import requests
from bs4 import  BeautifulSoup
import pandas as pd 
import re

                                         
```

```python
apartments_villas_for_sale=[]
ap=[]
for x in range(1,200):
    url=f'https://www.olx.com.lb/properties/apartments-villas-for-sale/?page={x}'
    base_url='https://www.olx.com.lb'
    r=requests.get(url)
    soup=BeautifulSoup(r.content, 'html.parser')
    lists=soup.find_all('ul', class_='ba608fb8 de8df3a3')
    for item in lists:
        link=item.find_all('div', class_='ee2b0479')
        for single in link:
            for i in single.find_all('a', href=True):
                ap.append(base_url+i['href'])


        
for ur in ap:
    r=requests.get(ur)
    soup=BeautifulSoup(r.content, 'html.parser')

    try:
        title=soup.find('h1', class_='a38b8112').text.strip()
    except AttributeError as err:
        title=''
    try:
        price=soup.find('div', class_='_1075545d').find('span', class_='_56dab877').text.strip()
    except AttributeError as err:
        price=''
    price_final=price.replace("USD", "")
    try:
        location=soup.find('div', class_='_1075545d e3cecb8b _5f872d11').find('span', class_='_8918c0a8').text.strip()
    except AttributeError as err:
        location=''
    try:
       space=soup.find('span',  attrs={'aria-label': 'Area'}).find('span', class_='c47715cd').text.strip() 
    except AttributeError as err:
            space=''
    try:
        seller=soup.find('div', class_='_1075545d _6caa7349 _42f36e3b d059c029').find('span', class_='_261203a9 _2e82a662').text.strip()
    except AttributeError as err:
        seller='-'
    try:
         since=soup.find('div', class_='_05330198').text.strip()
    except AttributeError as err:
        since='-'
    Member_since=since.replace('Member since', '')
    try:
        ad_id=soup.find('div', class_='_171225da').text.strip()
    except AttributeError as err:
        ad_id=' USD'
    id=ad_id.replace("Ad id","")
    try: 
        bedrooms=soup.select_one('span:contains("Bedrooms") + span').text.strip()
    except AttributeError as err:
        bedrooms='0'
    try:
        bathrooms=soup.select_one('span:contains("Bathrooms") + span').text.strip()
    except AttributeError as err:
        bathrooms='0'
    try:
        Area=soup.select_one('span:contains("Size") + span').text.strip()
    except  AttributeError as err:
        Area='-'
    try:
        Property_Type=soup.select_one('span:contains("Property Type") + span').text.strip()
    except AttributeError as err:
        Property_Type=''
    try:
        Payment_method=soup.select_one('span:contains("Payment method") + span').text.strip()
    except AttributeError as err:
        payment_method='Not mention'
    try:

        Condition=soup.select_one('span:contains("Condition") + span').text.strip()
    except AttributeError as err:  
        Condition=''
    try:
        Furnished=soup.select_one('span:contains("Furnished") + span').text.strip()
    except AttributeError as err:
        Furnished="Not mention"
    type = soup.find_all('li', class_='_8c543153')
    for t in type:
        s = t.find('a', href=True)
        types = s['title']
        
 
    apartments_villas_for_sale.append([id,title,location,price_final,bedrooms,bathrooms,Area,types,Property_Type,Payment_method,Condition,Furnished,seller,Member_since,ur])
   
print(apartments_villas_for_sale)

df=pd.DataFrame(apartments_villas_for_sale, columns=['id','title','location','price_final','bedrooms','bathrooms','Area','types','Property_Type','Payment_method','Condition','Furnished','seller','Member_since','link'])
```
